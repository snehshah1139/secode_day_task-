#**Step 3: Histograms + Boxplots**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")
numeric_cols = ['Population_Density', 'Nearby_Roads', 'Green_Cover_%', 'Noise_dB', 'Noise_Complaints']

# Create histograms and boxplots side by side
for col in numeric_cols:
    plt.figure(figsize=(12, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Histogram of {col}")

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()
```

# **Histograms + Boxplots: A Simplified Overview Throgh the Graph**

# **1] Noise Complaints:**

**->Histogram:** Most noise complaints are between 0‚Äì20. The highest number of complaints (up to 160) occurred in the lower ranges (e.g., 0‚Äì5).

**->Boxplot:** The majority of complaints are clustered in the lower range, with a few extreme cases (possibly very high complaints in some areas).

---------------------


# **2] Noise Levels (dB):**

**->Histogram:** Noise levels range from 20 dB (quiet) to 100 dB (very loud). Most areas have moderate noise (around 40‚Äì70 dB).

**->Boxplot:** Most data is in the middle range, with some very loud areas (up to 100 dB).

---------------------

# **3] Population Density:**

**->Histogram**: Density ranges from 0‚Äì10,000 people. Some areas are sparsely populated (low numbers), while others are very crowded (high numbers).

**->Boxplot:** The spread is wide, meaning some regions are much denser than others.


---------------------

# **4] Nearby Roads:**

**->Histogram**: Most places have 0‚Äì2 roads nearby. A few areas have up to 8 roads, but these are rare.

**->Boxplot:** Roads are mostly concentrated in the lower range (0‚Äì4), with a few exceptions.

# **# In Short:**
**The data shows patterns of noise, population, and roads. Most complaints and noise levels are low to moderate, population density varies greatly, and roads are mostly few but occasionally clustered.**

# **Step 5 ‚Äì Pairplot + Correlation Matrix**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns
numeric_cols = ['Population_Density', 'Nearby_Roads', 'Green_Cover_%', 'Noise_dB', 'Noise_Complaints']

# 1. Correlation matrix
corr = df[numeric_cols].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("üîó Correlation Matrix")
plt.show()

# 2. Pairplot (may take time if data is large)
sns.pairplot(df[numeric_cols])
plt.suptitle("üìà Pairwise Feature Relationships", y=1.02)
plt.show()
```

# **EDA Summary: Correlation + Pairwise Relationships Through The Upper Graphs**

1]Correlation Matrix (Color Grid):
---------------------------------
This heatmap shows how strongly two features are related, using numbers between -1 (opposite) and 1 (strong direct relation):

-Noise_dB ‚Üî Noise_Complaints ‚Üí Strong positive correlation (0.77)
‚û§ Meaning: More noise levels = More complaints

-All other feature combinations (e.g., Population, Green Cover, Roads)
‚Üí Have very weak or no correlation (close to 0)
‚û§ They don‚Äôt influence each other much


---------------------------------

2]Pairplot (Scatterplots + Histograms):
---------------------------------
